### Implementación Interpolación Numérica

Utilizaremos a modo de ejemplo:

$$
\int_0^1 x^2dx = 1/3
$$

```python
# Reglas incluyendo interpolación Compuesta:
def punto_medio(f, a, b, n):
    h = (b - a) / n
    suma = 0.0
    for i in range(n):
        x_med = a + (i + 0.5) * h
        suma += f(x_med)
    return h * suma


def trapecio(f, a, b, n):
    h = (b - a) / n
    suma = 0.5 * (f(a) + f(b))
    for i in range(1, n):
        x = a + i * h
        suma += f(x)
    return h * suma


def simpson(f, a, b, n):
    if n % 2 != 0:
        raise ValueError("n debe ser par para Simpson")
    
    h = (b - a) / n
    suma = f(a) + f(b)
    
    for i in range(1, n):
        x = a + i * h
        if i % 2 == 0:
            suma += 2 * f(x)
        else:
            suma += 4 * f(x)
    
    return h * suma / 3
```

```python
def f(x):
    return x**2

a = 0
b = 1
n = 100

print("Punto medio:", punto_medio(f, a, b, n))
print("Trapecio:", trapecio(f, a, b, n))
print("Simpson:", simpson(f, a, b, n))
```
